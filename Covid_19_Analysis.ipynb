```python
df = pd.read_csv("owid-covid-data.csv", parse_dates=['date'])
india = df[df['location'] == 'India']
```

```python
df.shape
```

```python
#specifically looking at the data of india
india
```

New and Total Cases

```python
plt.figure(figsize=(12,6))
plt.plot(india.date, india.new_cases_smoothed, label='New Cases (7‑day avg)')
plt.plot(india.date, india.total_cases, label='Total Cases', alpha=0.6)
plt.title("COVID‑19 in India: New & Total Cases")
plt.legend()
plt.show()
```

Deaths and Case Fatality Rate

```python
india['cfr'] = india.total_deaths / india.total_cases * 100
plt.figure(figsize=(12,6))
plt.plot(india.date, india.cfr)
plt.title("Case Fatality Rate (CFR) in India (%)")
plt.show()
```

Vaccination Progress


```python
plt.figure(figsize=(12,6))
plt.plot(india.date, india.people_vaccinated, label='At Least 1 Dose')
plt.plot(india.date, india.people_fully_vaccinated, label='Fully Vaccinated')
plt.title("Vaccination Trends in India")
plt.legend()
plt.show()
```

Testing vs Positivity

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=india.new_tests_smoothed_per_thousand,
                y=india.new_cases_smoothed_per_million,
                hue=india.date.dt.year)
plt.title("India: Tests vs Case Rates (7‑day smoothed)")
plt.show()
```

Correlation Heatmap

```python
num = india[['new_cases_smoothed', 'new_deaths_smoothed',
             'new_tests_smoothed', 'people_fully_vaccinated']].dropna()
plt.figure(figsize=(8,6))
sns.heatmap(num.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlations (India)")
plt.show()
```
